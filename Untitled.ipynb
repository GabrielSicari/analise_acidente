```python
# Análise exploratória de acidentes rodoviários (versão em Python)
# O código foi escrito para ser robusto ao arquivo resumido enviado (Book1.xlsx)
# e também funcionar em uma versão maior do mesmo dataset.

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

# ==========================
# 1) Carregamento e preparo
# ==========================
file_path = 'acidentes_brasil - acidentes_brasil.csv'

df = pd.read_csv(file_path)
original_cols = df.columns.tolist()

# Converter tipos e criar colunas derivadas
# Datas
if 'data_inversa' in df.columns:
    df['data'] = pd.to_datetime(df['data_inversa'], errors='coerce')
else:
    # fallback se o nome vier diferente
    date_like = [c for c in df.columns if 'data' in c.lower()]
    df['data'] = pd.to_datetime(df[date_like[0]], errors='coerce') if date_like else pd.NaT

# Hora
if 'horario' in df.columns:
    df['hora'] = pd.to_datetime(df['horario'].astype(str), errors='coerce', format='%H:%M:%S').dt.hour
else:
    df['hora'] = np.nan

# Ano, Mês (período), Chave ano-mês
df['ano'] = df['data'].dt.year
df['mes_periodo'] = df['data'].dt.to_period('M')
df['ano_mes'] = df['mes_periodo'].astype(str)

# Normalizações numéricas
for col in ['km','mortos','feridos_leves','feridos_graves','veiculos','pessoas','ilesos','ignorados','feridos','latitude','longitude']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Total de feridos (se existir leve e grave)
if 'feridos_leves' in df.columns or 'feridos_graves' in df.columns:
    df['total_feridos'] = df.get('feridos_leves', 0).fillna(0) + df.get('feridos_graves', 0).fillna(0)
else:
    df['total_feridos'] = df.get('feridos', 0).fillna(0)

# Flag de acidente fatal
# Considera "mortos" > 0 e/ou texto "Fatais" na classificação, quando existir
class_col = 'classificacao_acidente' if 'classificacao_acidente' in df.columns else None
if 'mortos' in df.columns:
    mortos_pos = (df['mortos'].fillna(0) > 0)
else:
    mortos_pos = pd.Series(False, index=df.index)

if class_col:
    fatais_por_texto = df[class_col].astype(str).str.contains('Fatais', case=False, na=False)
else:
    fatais_por_texto = pd.Series(False, index=df.index)

df['fatal'] = (mortos_pos | fatais_por_texto).astype(int)

# Dia da semana (normalização leve para ordenação)
if 'dia_semana' in df.columns:
    d = df['dia_semana'].astype(str).str.lower().str.strip()
    map_day = {
        'segunda-feira': 'segunda',
        'terca-feira': 'terca',
        'terça-feira': 'terca',
        'quarta-feira': 'quarta',
        'quinta-feira': 'quinta',
        'sexta-feira': 'sexta',
        'sabado': 'sabado',
        'sábado': 'sabado',
        'domingo': 'domingo',
    }
    df['dia_semana_simpl'] = d.map(map_day).fillna(d)
    ordem_semana = ['segunda','terca','quarta','quinta','sexta','sabado','domingo']
    df['dia_semana_simpl'] = pd.Categorical(df['dia_semana_simpl'], categories=ordem_semana, ordered=True)
else:
    df['dia_semana_simpl'] = np.nan

# Assegurar string limpa para UF e outros campos-chave
for c in ['uf','municipio','condicao_metereologica','tipo_pista','tracado_via','sentido_via','fase_dia','tipo_acidente','causa_acidente']:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# ===================================
# 2) Sumário geral e estatísticas
# ===================================
qtd_registros = len(df)
qtd_ufs = df['uf'].nunique() if 'uf' in df.columns else np.nan
qtd_municipios = df['municipio'].nunique() if 'municipio' in df.columns else np.nan

acidentes_fatais = int(df['fatal'].sum())
vitimas_fatais = int(df['mortos'].fillna(0).sum()) if 'mortos' in df.columns else 0
feridos_total = int(df['total_feridos'].fillna(0).sum())

print('=== Sumário Geral ===')
print(f'Registros: {qtd_registros}')
print(f'UFs distintas: {qtd_ufs}')
print(f'Municípios distintos: {qtd_municipios}')
print(f'Acidentes fatais: {acidentes_fatais}')
print(f'Vítimas fatais (soma de "mortos"): {vitimas_fatais}')
print(f'Feridos (leves+graves): {feridos_total}')

# ===================================
# 3) Análise Temporal
# ===================================
if df['data'].notna().any():
    # Acidentes por ano-mês
    am = df.groupby('mes_periodo').size().rename('acidentes').reset_index()
    am = am.sort_values('mes_periodo')
    if len(am) > 0:
        plt.figure(figsize=(10,4))
        plt.plot(am['mes_periodo'].dt.to_timestamp(), am['acidentes'], marker='o')
        plt.title('Acidentes por mês (ano-mês)')
        plt.xlabel('Ano-Mês')
        plt.ylabel('Quantidade de acidentes')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

    # Acidentes por dia da semana
    if df['dia_semana_simpl'].notna().any():
        sem = df.groupby('dia_semana_simpl').size().rename('acidentes').reset_index()
        plt.figure(figsize=(8,4))
        sns.barplot(data=sem, x='dia_semana_simpl', y='acidentes', color='#4C72B0')
        plt.title('Acidentes por dia da semana')
        plt.xlabel('Dia da semana')
        plt.ylabel('Quantidade de acidentes')
        plt.tight_layout()
        plt.show()

    # Acidentes por hora do dia
    if df['hora'].notna().any():
        horas = df.groupby('hora').size().rename('acidentes').reset_index()
        plt.figure(figsize=(10,4))
        sns.barplot(data=horas, x='hora', y='acidentes', color='#55A868')
        plt.title('Acidentes por hora do dia')
        plt.xlabel('Hora')
        plt.ylabel('Quantidade de acidentes')
        plt.tight_layout()
        plt.show()

# ===================================
# 4) Análise Geográfica
# ===================================
if 'uf' in df.columns:
    # Top UFs por acidentes e taxa de fatalidade
    uf_stats = df.groupby('uf').agg(
        acidentes=('fatal', 'size'),
        acidentes_fatais=('fatal','sum'),
        mortos=('mortos','sum') if 'mortos' in df.columns else ('fatal','sum')
    ).reset_index()
    uf_stats['taxa_fatalidade_%'] = (100*uf_stats['acidentes_fatais']/uf_stats['acidentes']).round(2)
    uf_stats = uf_stats.sort_values(['acidentes_fatais','acidentes'], ascending=[False, False])
    print('\n=== Top UFs por acidentes e fatalidade ===')
    print(uf_stats.head(15))

    plt.figure(figsize=(10,4))
    top_ufs = uf_stats.sort_values('acidentes', ascending=False).head(10)
    sns.barplot(data=top_ufs, x='uf', y='acidentes', color='#4C72B0')
    plt.title('Top 10 UFs por quantidade de acidentes')
    plt.xlabel('UF')
    plt.ylabel('Acidentes')
    plt.tight_layout()
    plt.show()

# Scatter geográfico simples (sem mapa base), destacando fatais
if {'latitude','longitude'}.issubset(df.columns):
    geo = df.dropna(subset=['latitude','longitude'])
    if len(geo) > 0:
        plt.figure(figsize=(6,6))
        plt.scatter(geo['longitude'], geo['latitude'], s=10, alpha=0.25, label='Acidente')
        fatais_geo = geo[geo['fatal']==1]
        if len(fatais_geo) > 0:
            plt.scatter(fatais_geo['longitude'], fatais_geo['latitude'], s=18, alpha=0.8, c='red', label='Fatal')
        plt.title('Distribuição geográfica (lon x lat)')
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')
        plt.legend()
        plt.tight_layout()
        plt.show()

# ===================================
# 5) Condicionantes (clima, pista, traçado, fase do dia)
# ===================================

def plot_factor_rate(df, factor_col, top_n=10, titulo_prefix=''):
    if factor_col not in df.columns:
        return
    aux = df.copy()
    aux[factor_col] = aux[factor_col].fillna('Ignorado')
    g = aux.groupby(factor_col).agg(
        acidentes=('fatal','size'),
        fatais=('fatal','sum'),
        mortos=('mortos','sum') if 'mortos' in aux.columns else ('fatal','sum')
    ).reset_index()
    g['taxa_fatalidade_%'] = (100*g['fatais']/g['acidentes']).round(2)
    g = g.sort_values('acidentes', ascending=False).head(top_n)
    print(f"\n=== {titulo_prefix}{factor_col}: acidentes, fatais, taxa(%) ===")
    print(g)
    # Plot
    fig, ax1 = plt.subplots(figsize=(10,4))
    sns.barplot(data=g, x=factor_col, y='acidentes', color='#4C72B0', ax=ax1)
    ax1.set_xlabel(factor_col)
    ax1.set_ylabel('Acidentes')
    ax1.set_title(f'{titulo_prefix}{factor_col}: Top {top_n} por acidentes')
    ax1.tick_params(axis='x', rotation=35)
    ax2 = ax1.twinx()
    ax2.plot(range(len(g)), g['taxa_fatalidade_%'], color='#C44E52', marker='o', label='Taxa fatal (%)')
    ax2.set_ylabel('Taxa fatal (%)')
    fig.tight_layout()
    plt.show()

for col, title in [
    ('condicao_metereologica','Condição: '),
    ('tipo_pista','Tipo de pista: '),
    ('tracado_via','Traçado da via: '),
    ('fase_dia','Fase do dia: '),
    ('tipo_acidente','Tipo de acidente: '),
    ('causa_acidente','Causa do acidente: '),
]:
    plot_factor_rate(df, col, top_n=10, titulo_prefix=title)

# ===================================
# 6) Comparações Regionais e Temporais
# ===================================
if 'uf' in df.columns and df['data'].notna().any():
    comp = df.groupby(['uf','mes_periodo']).agg(
        acidentes=('fatal','size'),
        fatais=('fatal','sum'),
        mortos=('mortos','sum') if 'mortos' in df.columns else ('fatal','sum')
    ).reset_index()
    comp['taxa_fatal_%'] = (100*comp['fatais']/comp['acidentes']).round(2)
    print('\n=== Comparação UF x Mês (amostra) ===')
    print(comp.sort_values(['uf','mes_periodo']).head(30))

    # Pequeno painel: para as top 6 UFs por acidentes totais
    top6_ufs = (
        comp.groupby('uf')['acidentes'].sum().sort_values(ascending=False).head(6).index.tolist()
    )
    comp_top6 = comp[comp['uf'].isin(top6_ufs)].copy()
    if len(comp_top6) > 0:
        g = sns.relplot(
            data=comp_top6,
            x='mes_periodo', y='acidentes', hue='uf', kind='line', height=4, aspect=1.6, marker='o'
        )
        g.set_axis_labels('Ano-Mês','Acidentes')
        plt.title('Acidentes por mês - Top 6 UFs (amostra)')
        for ax in g.axes.flat:
            ax.tick_params(axis='x', rotation=45)
        plt.tight_layout()
        plt.show()

# ===================================
# 7) Identificação de Trechos Críticos
# ===================================
# Agrupar por UF, BR, Km e priorizar por acidentes fatais e mortos
keys = [k for k in ['uf','br','km'] if k in df.columns]
if len(keys) == 3:
    # Arredondar KM para reduzir ruído (opcional). Mantendo como está para maior precisão.
    black = df.groupby(keys).agg(
        acidentes=('fatal','size'),
        fatais=('fatal','sum'),
        mortos=('mortos','sum') if 'mortos' in df.columns else ('fatal','sum'),
        feridos_graves=('feridos_graves','sum') if 'feridos_graves' in df.columns else ('fatal','sum'),
        feridos_leves=('feridos_leves','sum') if 'feridos_leves' in df.columns else ('fatal','sum')
    ).reset_index()
    # Ordenar por mais críticos: fatais, mortos, acidentes
    black = black.sort_values(['fatais','mortos','acidentes'], ascending=[False, False, False])
    # Filtrar para exibir apenas trechos com alguma gravidade ou mais de 1 acidente
    black_filtrado = black[(black['fatais']>0) | (black['acidentes']>=2)]
    print('\n=== Trechos críticos (UF, BR, Km) — top 20 ===')
    print(black_filtrado.head(20))

    # Visual rápido: Top 10 por acidentes fatais
    top10_black = black_filtrado.head(10)
    if len(top10_black) > 0:
        plt.figure(figsize=(10,5))
        labels = top10_black.apply(lambda r: f"{r['uf']}-BR{int(r['br']) if pd.notna(r['br']) else 'NA'} @km {r['km']}", axis=1)
        sns.barplot(x=labels, y=top10_black['fatais'], color='#C44E52')
        plt.title('Top 10 trechos por acidentes fatais')
        plt.xlabel('Trecho (UF-BR @km)')
        plt.ylabel('Acidentes fatais')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

# =====================
# 8) Observações finais
# =====================
print('\nOBS: Esta análise reflete a amostra carregada. Em um dataset completo (mais datas/UFs), as tendências temporais, comparações regionais e identificação de hotspots ficam mais robustas.\n')
```
